All the data that shared is one file per given year. To concat everything together, create dataframe for each dataset into array of dataframe and concat using pandas.concat

```python
years = range(1880,2011)
pieces = []
columns = ['name','sex','births']

for year in years:
    path = 'ch02/names/yob%d.txt'% year
    
    frame = pd.read_csv(path,names=columns)
    frame['year'] = year
    pieces.append(frame)
```

```python
len(pieces)
```

```python
names = pd.concat(pieces,ignore_index=True)#use ignore index so we don't preserve original index for each dataframe
```

```python
names.tail()
```

```python
total_births = names.pivot_table('births',rows='year',columns='sex',aggfunc='sum')
```

```python
total_births.head()
```

```python
%pylab inline
```

```python
total_births.plot(title='Total births by sex and year')
```

```python
def add_prop(group):
    births = group.births.astype(float)
    
    group['prop'] = births / births.sum()
    return group
```

```python
names = names.groupby(['year','sex']).apply(add_prop)
```

```python
names.tail()
```

```python
#perform sanity check if all operation close to but not same to one

np.allclose(names.groupby(['year','sex']).prop.sum(),1)
```

```python
def top_1000(group):
    return group.sort_index(by='births',ascending=False)[:1000]
```

```python
top1000 = names.groupby(['year','sex']).apply(top_1000)
```

```python
top1000.head()
```

```python
boys = top1000[top1000.sex == 'M']
girls = top1000[top1000.sex == 'F']
```

```python
total_births = top1000.pivot_table('births',rows='year',columns='name',aggfunc='sum')
```

The decreasing plot can be mean as fewer parents choosing common names. Let's validate this hypothesis

```python
table = top1000.pivot_table('prop', rows='year',columns='sex',aggfunc='sum')
table.plot(title='Sum of top 1000 name of prop by year and sex',
           yticks = np.linspace(0,1.2,13),
           xticks = range(1880,2020,10))
```

You can see that both male and female has been decreasing, by sum up the propotion of top 1000 names. This validate the hypothesis earlier that parents begin searching for more unique names.

```python
df.head()
```

We want to know 50% of people with the most unique names. It's tricky to do this task. To do this, we need to know what index to stop in when we have achieve 50% people. We can do this by cumulative sum, np.sum(). Then we need to know what index if we want to insert 0.5 while keeping sorted order. Do this by searchsorted

```python
prop_cumsum = df.sort_index(by='prop',ascending=False).prop.cumsum()
```

```python
prop_cumsum.head()
```

```python
prop_cumsum.values.searchsorted(0.5)
```

Compare to 1900, where the number is much smaller.

```python
df = boys[boys.year == 1900]
in1900 = df.sort_index(by='prop',ascending=False).prop.cumsum()
in1900.values.searchsorted(0.5) + 1
```

After we now how to return count sorted for each specific sex and specific year, we can perform this function by using our whole data groupby year and sex. This will return groupby, and using aggfunc of get_count_group and use those values.

```python
def get_count_group(group,q=0.5):
    group = group.sort_index(by='prop',ascending=False).prop.cumsum()
    return group.values.searchsorted(q)+1

diversity = top1000.groupby(['year','sex']).apply(get_count_group)
```

```python
diversity.head()
```

```python
diversity = diversity.unstack('sex')
```

```python
diversity.head()
```

```python
diversity.plot(title='number of top popular names in top 50%')
```
